```python
import requests
import pandas as pd
from lxml import etree
import time
from random import randint
```

```python
ROOT_URL = 'https://www.franceculture.fr/emissions/les-chemins-de-la-philosophie?p={}'
range_url = range(1,5)
```

```python
ROOT_URL.format(1)
```

```python
def get_list_of_podscast_on_page(page_number):
    url =ROOT_URL.format(str(page_number))
    print(url)
    r = requests.get(url)
    e = etree.HTML(r.content)
    list_dict = []
    elements = e.xpath("//div[@class='teaser-replay-button-wrapper']//button")
    #return elements
    list_dic  = [dict(d.attrib) for d in elements]
    return pd.DataFrame(list_dic)
```

```python
def retrieve_all_podcasts_metadata():
    df_all = pd.DataFrame()
    for i in range_url:
        try:
            df = get_list_of_podscast_on_page(i)
            df_all = pd.concat([df_all,df],axis=0)
        except Exception as e:
            print(f'shoundi, this url did not work : {str(i)}')
            print(e)
        finally:
            time.sleep(1+randint(1,4))
        
    df_all.to_csv('all_podcasts_1.csv')
```

```python
retrieve_all_podcasts_metadata()
```
